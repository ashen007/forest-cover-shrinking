```python
import os
import glob
import rasterio
from rasterio.plot import show

import pandas as pd
```

```python
root = '../../data/HRSCD/'
change_map = os.path.join(root, 'change')
years = ['2006', '2012']
subset = 'D14'
```

```python
image_df = pd.DataFrame()
image_df["2006"] = os.listdir(os.path.join(root, years[1], subset))
image_df["change"] = os.listdir(os.path.join(change_map, subset))

image_df.head(10)
```

```python
file_name = image_df.loc[0, "change"]
path_06 = os.path.join(root, years[0], subset, file_name)
path_12 = os.path.join(root, years[1], subset, file_name)
path_change = os.path.join(change_map, subset, file_name)

path_06, path_12, path_change
```

```python
with rasterio.open(path_06) as image:
    img_arr_06 = image.read()
    
with rasterio.open(path_12) as image:
    img_arr_12 = image.read()
    
with rasterio.open(path_change) as image:
    img_arr_change = image.read()
```

```python
show(img_arr_06, cmap='bone_r')
```

```python
show(img_arr_12, cmap='bone_r')
```

```python
show(img_arr_change, cmap='bone_r')
```
